### Ocena z ćwiczeń

Na ocenę z ćwiczeń będą miały wpływ następujące czynniki:
1. Zadania domowe (40%)
2. Mini-kolokwia (30%)
3. Projekty (30%)


#### Zadania domowe
1. W trakcie zajęć studenci otrzymają do wypełnienia notebook z ćwiczeniami. Notebook z danych zajęć musi być uzupełniony najpóźniej przed rozpoczęciem kolejnych zajęć. Przed kolejnymi zajęciami Studenci deklarują, które z zadanych zadań rozwiązali poprawnie w udostępnionym docsie. 
2. W przypadku ćwiczeń tablicowych studenci otrzymają w wyprzedzeniem listę zadań opublikowaną w repozytorium przedmiotu oraz na Pegazie. W przypadku prezentacji zadania tablicowego nie dopuszczam możliwości przepisania całego zadania z notatek.
3. Na początku każdych zajęć prowadzący będzie prosić o przedstawienie i omówienie danego zadania, które zostało zadane na poprzednich zajęciach. W przypadku niezrozumienia przez studenta fragmentu lub całości kodu, istnieje możliwość odjęcia punktów. 
4. Za każdy poprawnie uzupełniony notebook można dostać maksymalnie 10 punktów.

Osoby dołączające do zajęć w późniejszym terminie muszą liczyć się z faktem, iż nie mogą otrzymać punktów za zaległe zadania.

#### Mini - kolokwia
W trakcie semestru planowane są trzy mini-kolokwia. Termin planowanego mini-kolokwium będzie zapowiedziany z conajmniej tygodniowym wyprzedzeniem. Ponadto mini-kolokwia będą przeprowadzone stacjonarnie w sposób tradycyjny (na kartkach) bez dostępu do komputera.

#### Projekty
W ramach zaliczenia przedmiotu każda grupa będzie pracować nad projektem. Szczegółowe wymagania oraz kryteria oceny zostaną omówione na ćwiczeniach.

#### Pozostałe
Do rozwiązania zadań dopuszczam korzystanie z LLMów, ale jednocześnie wymagam dokładnego zrozumienia wygenerowanych rozwiązań.

---
Ocena z ćwiczeń będzie obliczana zgodnie z klasyczną skalą ocen, tj.: 

* [0-50) %: 2
* [50-60) %: 3
* [60-70) %: 3,5
* [70-80) %: 4
* [80-90) %: 4,5
* [90-100) %: 5
